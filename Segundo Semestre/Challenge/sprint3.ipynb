# Algoritmo de Busca

## Merge Sort

```python
# Funcao para para definir parametros iniciais
def merge_sort(lista, key, reverse=False):
    # Se a lista só tiver um elemento ou estiver vazia, já está ordenada
    if len(lista) <= 1:
        return lista

    # Definir o meio. Divide a lista em duas partes: esquerda e direita
    meio = len(lista) // 2
    esquerda = merge_sort(lista[:meio], key, reverse)
    direita = merge_sort(lista[meio:], key, reverse)

    # Chama a próxima função já com os parâmetros definidos
    return merge(esquerda, direita, key, reverse)

# Função para mesclar as duas listas ordenadas
def merge(esquerda, direita, key, reverse):
    resultado = []
    i = j = 0

    # Enquanto houver elementos em ambas as listas, compara e adiciona o menor (ou maior) à lista resultado
    while i < len(esquerda) and j < len(direita):
        valor_esq = esquerda[i][key]
        valor_dir = direita[j][key]

        # Se ordenação for decrescente (reverse=True)
        if reverse:
            if valor_esq > valor_dir:
                resultado.append(esquerda[i])
                i += 1
            else:
                resultado.append(direita[j])
                j += 1
        # Se ordenação for crescente (reverse=False)
        else:
            if valor_esq < valor_dir:
                resultado.append(esquerda[i])
                i += 1
            else:
                resultado.append(direita[j])
                j += 1

    # Adiciona os elementos restantes (caso uma das listas acabe antes da outra)
    resultado.extend(esquerda[i:])
    resultado.extend(direita[j:])
    return resultado

# Chave para Ordenação
chave_para_ordenar = input("Digite a chave para ordenar (ex: preco, quantidade, categoria): ")

# Verifica se a chave existe no dicionário
if chave_para_ordenar not in banco_dados[0]:
    print("Chave inválida. Tente novamente com uma chave existente.")
else:
    banco_crescente = merge_sort(banco_dados, chave_para_ordenar)
    banco_decrescente = merge_sort(banco_dados, chave_para_ordenar, reverse=True)

# Transformando em df para melhorar a visualização
df_crescente = pd.DataFrame(banco_crescente)
df_decrescente = pd.DataFrame(banco_decrescente)
```

## Quick Sort

```python
# Funcao para definir parametros iniciais
def quick_sort(lista, key, reverse=False):
    # Se a lista só tiver um elemento ou estiver vazia, já está ordenada
    if len(lista) <= 1:
        return lista

    # Definir primeiro elemento como pivo. Listas para armazenar elementos menores, iguais e maiores
    pivo = lista[0]
    menores = []
    iguais = []
    maiores = []

    for item in lista:
        # Selecionar os valores de acordo com a categoria selecionada
        valor = item[key]
        pivo_valor = pivo[key]

        # Definir valores menores, iguais e maiores ao pivo e armazenar
        # menores: valores menores que o pivô (ou maiores se reverse=True)
        # iguais: valores iguais ao pivô
        # maiores: valores maiores que o pivô (ou menores se reverse=True)
        if valor == pivo_valor:
            iguais.append(item)
        elif (valor < pivo_valor and not reverse) or (valor > pivo_valor and reverse):
            menores.append(item)
        else:
            maiores.append(item)
    
    # Retornar a própria função para ordenar as sublistas e concatenação do resultado final
    return quick_sort(menores, key, reverse) + iguais + quick_sort(maiores, key, reverse)

# Chave para Ordenação
chave_para_ordenar = input("Digite a chave para ordenar (ex: preco, quantidade, categoria): ")

# Verifica se a chave existe no dicionário
if chave_para_ordenar not in banco_dados[0]:
    print("Chave inválida. Tente novamente com uma chave existente.")
else:
    banco_crescente = quick_sort(banco_dados, chave_para_ordenar)
    banco_decrescente = quick_sort(banco_dados, chave_para_ordenar, reverse=True)

# Transformando em df para melhorar a visualização
df_crescente = pd.DataFrame(banco_crescente)
df_decrescente = pd.DataFrame(banco_decrescente)
```

# Ordenação Parcial

```python
import heapq

def k_maiores(dados, chave, k):
    return heapq.nlargest(k, dados, key=lambda x: x[chave])

def k_menores(dados, chave, k):
    return heapq.nsmallest(k, dados, key=lambda x: x[chave])


top3_precos = k_maiores(banco_dados, "preco", 3)
menor_estoque = k_menores(banco_dados, "quantidade", 2)
```

# Árvore de Pequisa Binária

```python
# Classe que representa um nó da Árvore Binária de Busca (BST)
class NoBST:
    def __init__(self, chave, valor):
        self.chave = chave         # Valor que será usado para ordenar os nós (peca, preco, quantidade etc.)
        self.valor = valor         # Valor completo (ex: o dicionário inteiro com os dados da peca)
        self.esquerda = None       # Subárvore esquerda (valores menores)
        self.direita = None        # Subárvore direita (valores maiores)

# Função para inserir um novo nó na BST
def inserir_bst(raiz, chave, valor):
    # Se a árvore está vazia, cria um novo nó
    if raiz is None:
        return NoBST(chave, valor)
    
    # Se a chave for menor, insere na subárvore esquerda. Se for maior, insere na subárvore direita
    # Se a chave já existir, não faz nada (não insere duplicados)
    if chave < raiz.chave:
        raiz.esquerda = inserir_bst(raiz.esquerda, chave, valor)
    else:
        raiz.direita = inserir_bst(raiz.direita, chave, valor)
    
    # Retorna a raiz da árvore (ou subárvore) após a inserção
    return raiz

# Função para buscar sugestões de autocompletar com base em um prefixo
def buscar_prefixo(raiz, prefixo, resultados):
    # Se a árvore está vazia, retorna
    if raiz is None:
        return
    
    # Se a chave do nó atual começa com o prefixo, adiciona o valor à lista de resultados
    if raiz.chave.startswith(prefixo):
        resultados.append(raiz.valor)

    # Continua a busca na subárvore esquerda e direita
    buscar_prefixo(raiz.esquerda, prefixo, resultados)
    buscar_prefixo(raiz.direita, prefixo, resultados)

# Função principal de autocomplete (busca por prefixo com limite de resultados)
def autocomplete(bst, prefixo, limite=5):
    resultados = []
    buscar_prefixo(bst, prefixo, resultados)
    return resultados[:limite]

# Construindo a árvore binária de busca com base nas peças
bst_pecas = None
for item in banco_dados:
    bst_pecas = inserir_bst(bst_pecas, item['peca'], item)

# Função para buscar sugestões
def buscar_sugestoes(bst, limite=5):
    prefixo = input("Digite o prefixo para busca: ")
    resultados = autocomplete(bst, prefixo, limite)
    if resultados:
        print(f"Sugestões para '{prefixo}':")
        for item in resultados:
            print(f"{item['peca']} - {item['modelo_carro']}")
    else:
        print(f"Nenhuma sugestão encontrada para '{prefixo}'.")

# Chamar a função de busca
buscar_sugestoes(bst_pecas)
```
